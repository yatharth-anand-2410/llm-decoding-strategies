# 12. DoLa (Decoding by Contrasting Layers)
DoLa improves factuality by subtracting the logits of a premature layer from the mature final layer.

```python
import numpy as np

def dola_score(logits_mature, logits_premature, penalty=1.0):
    # Simple dynamic contrast
    return logits_mature - penalty * logits_premature
```
