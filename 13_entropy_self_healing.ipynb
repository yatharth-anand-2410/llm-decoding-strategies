# 13. Entropy-Triggered Self-Healing
Monitor entropy in real-time. If the model starts looping, entropy plummets. We can detect this and switch to Contrastive Search.

```python
import numpy as np

def calculate_entropy(probs):
    # Avoid log(0)
    p_safe = probs[probs > 0]
    return -np.sum(p_safe * np.log(p_safe))
```
